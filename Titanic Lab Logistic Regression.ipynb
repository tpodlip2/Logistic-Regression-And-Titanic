```python
X = pd.DataFrame()
X['sex'] = df['Sex']
X['age'] = df['Age']
X['pclass'] = df['Pclass']
X['sibsp'] = df['SibSp']
X['parch'] = df['Parch']
X['survived'] = df['Survived']
```

Looks like there are no null values for sex, siblings/spouses, or passenger class.  However, we see that there are null values for age.  This is not a good thing and unlike the lecture, I do not want to just drop these null values.  They will need to be filled.

```python
y = X['survived']
X = X.drop(['survived'], axis = 1)
```

```python
X['sex'] = pd.get_dummies(X.sex)['female']
```

Above we see that we have created a df of a single column of our survived information.  We also drop that column from our X df.  Then we take sex and change it from a categorical var to a numerical one where male and female will correspond to ones and zeroes. 

From this paper "The RMS Titanic - Machine Learning from Disaster", I see that the authors take the easier route of using the median age with regard to both sex and passenger class to fill in the missing age values.  I attempted doing a linear regression and failed! I still want to try it, but since I am low on time, this is the only path that worked for me.  I still think I have a good idea of how to have a wicked AUC.  Anyway, I liked the authors' simple approach.  

```python
agefill = numpy.zeros((2,3))
```

```python
for i in range (0,2):
    for j in range(0,3):
        agefill[i,j] = X[(X['sex'] == i) &
                         (X['pclass'] == j+1)]['age'].dropna().median()
```

```python
for i in range (0,2):
    for j in range(0,3):
        X.loc[(X['age'].isnull()) & (X['sex'] == i) & (X['pclass'] == j+1), 'age'] = agefill[i,j]
```
